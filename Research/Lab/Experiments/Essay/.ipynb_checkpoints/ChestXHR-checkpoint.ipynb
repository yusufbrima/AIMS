```python
os.mkdir('dataset')
train_path = '/content/dataset/train'
test_path = '/content/dataset/test'
path  =  '/content/dataset/'
os.mkdir(train_path)
os.mkdir(test_path)
base = '/content/COVID-19_Radiography_Dataset'
paths   = os.listdir(base)
```

```python
# We are creating the train and test directories
for k in range(len(paths)):
  if os.path.isdir(os.path.join(base,paths[k])):
    os.mkdir(os.path.join(train_path,paths[k]))
    os.mkdir(os.path.join(test_path,paths[k]))
```

```python
# We are creating our test set
if os.path.isdir(path) is True:
  base = '/content/COVID-19_Radiography_Dataset'
  paths   = os.listdir(base)
  pr =  0.2
  for k in range(len(paths)):
    if os.path.isdir(os.path.join(base,paths[k])):
      # We are moving 2% of the sample dataset to test

      n = np.int(len(os.listdir(os.path.join(base,paths[k]))) * pr)
      # print(random.sample(glob.glob(os.path.join(base,paths[k]+'/'+paths[k]+'*')),n))
      for c in random.sample(glob.glob(os.path.join(base,paths[k]+'/'+paths[k]+'*')),n):
        shutil.move(c,os.path.join(test_path,paths[k]))
```

```python
# We are moving the remaining dataset to train folder

if os.path.isdir(path) is True:
  base = '/content/COVID-19_Radiography_Dataset'
  paths   = os.listdir(base)
  for k in range(len(paths)):
    if os.path.isdir(os.path.join(base,paths[k])):
      for file in os.listdir(os.path.join(base,paths[k])):
        shutil.move(os.path.join(os.path.join(base,paths[k]),file),os.path.join(train_path,paths[k]))
```

```python
CLASSES  = os.listdir(train_path) #['Normal', 'Lung_Opacity', 'COVID', 'Viral Pneumonia']

BATCH_SIZE  = 10
HEIGHT,WIDTH = 224,224


train_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input,
                                   validation_split=0.2)

val_datagen = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input, validation_split=0.2)


#load the images to training
train_batches = train_datagen.flow_from_directory(directory=train_path, 
                                              target_size=(HEIGHT,WIDTH),
                                              class_mode='categorical',
                                              subset='training',
                                              shuffle=True, classes=CLASSES,
                                              batch_size=BATCH_SIZE, 
                                              color_mode="rgb")
#load the images to test
valid_batches = val_datagen.flow_from_directory(directory=train_path, 
                                              target_size=(HEIGHT,WIDTH),
                                              class_mode='categorical',
                                              subset='validation',
                                              shuffle=False, classes=CLASSES,
                                              batch_size=BATCH_SIZE, 
                                              color_mode="rgb")

# train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input) \
#     .flow_from_directory(directory=train_path,subset='training', target_size=(HEIGHT,WIDTH), classes=CLASSES, batch_size=BATCH_SIZE)


# valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input) \
#     .flow_from_directory(directory=train_path,validation_split=0.3,subset='validation',shuffle=False, target_size=(HEIGHT,WIDTH), classes=CLASSES, batch_size=BATCH_SIZE)



test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.vgg16.preprocess_input) \
    .flow_from_directory(directory=test_path, target_size=(HEIGHT,WIDTH), classes=CLASSES, batch_size=BATCH_SIZE, shuffle=False)
```

```python
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 10, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip( images_arr, axes):
        ax.imshow(img,cmap=plt.cm.Blues)
        ax.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
imgs, labels = next(train_batches)
```

```python
plotImages(imgs)
```

```python
vgg16_model = tf.keras.applications.vgg16.VGG16()
```

```python
vgg16_model.summary()
```

```python
model =  Sequential()
for layer in vgg16_model.layers[:-1]:
  model.add(layer)
```

```python
model.summary()
```

```python
for layer in model.layers:
    layer.trainable = False
```

```python
model.summary()
```

```python
model.add(Dense(units=4, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(optimizer=Adam(learning_rate=0.0001), loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history  = model.fit(x=train_batches,
          steps_per_epoch=len(train_batches),
          validation_data=valid_batches,
          validation_steps=len(valid_batches),
          epochs=5,
          verbose=2
)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
predictions = model.predict(x=test_batches, steps=len(test_batches), verbose=0)
```

```python
cm = confusion_matrix(y_true=test_batches.classes, y_pred=np.argmax(predictions, axis=-1))
```
