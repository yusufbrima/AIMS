```python
x = np.linspace(-10,10,100)
y = np.sin(x) + np.random.normal(0,0.09,len(x))
fig= plt.figure()
axes= fig.add_axes([0,0,1,1])
y1 =  y -  np.std(y)
y2 =  y + np.std(y)
axes.plot(x,y, color='blue', alpha=0.5,linewidth=2.5,label="Sin(x)+Noise")
axes.fill_between(x, y1, y2,color="green",alpha=0.1)

plt.show()
```
