```python
import numpy as np
import matplotlib.pyplot as plt
import numdifftools as nd
```

```python
def f(x):
  return x**2  - x + 1


def f2(x):
  return 2*x**2 + 1


def f3(x):
  return 2*x**2  -  4*x  + np.exp(x) -1

def gd(x,alpha,epsilon,f,N=100):
  gradNorm =  np.linalg.norm(nd.Derivative(f)(x))
  while gradNorm > epsilon:
    gradNorm =  np.linalg.norm(nd.Derivative(f)(x))
    x  =  x -  alpha * gradNorm
  return x
```

```python

alpha  = 0.005 
x  =  5
epsilon  = 0.001
```

```python
x2  = gd(x,alpha,epsilon,f)
xdata  =  np.linspace(-10,10,100)
plt.plot(xdata,f(xdata),label="Actual")
plt.plot(x2,f(x2),'ro',label="Estimated Minimum")
plt.title("Gradient Descent")
plt.show()
```

```python
x2  = gd(x,alpha,epsilon,f2)
xdata  =  np.linspace(-10,10,100)
plt.plot(xdata,f2(xdata),label="Actual")
plt.plot(x2,f2(x2),'ro',label="Estimated Minimum")
plt.title("Gradient Descent")
plt.legend()
plt.show()
```

```python
x2  = gd(x,alpha,epsilon,f3)
xdata  =  np.linspace(-10,10,100)
plt.plot(xdata,f3(xdata),label="Actual")
plt.plot(x2,f3(x2),'ro',label="Estimated Minimum")
plt.title("Gradient Descent")
plt.legend()
plt.show()
```
