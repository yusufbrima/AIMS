```python
import numpy as np
from sympy import symbols, solve
import matplotlib.pyplot as plt
```

```python
x = symbols('x')
expr = 32*x**3 - 48*x**2 + 17.08*x - 0.54 
sol = solve(expr)
sol
```

```python
def func(x):
    return 32*x**3 - 48*x**2 + 17.08*x - 0.54 
x = np.linspace(-0.2,1.2)
f =func(x)
plt.plot(x,f)
plt.xlabel("X2")
plt.ylabel("f(x)")

plt.grid()
```

```python
def func(x):
    return (0.5 - x)**2 + (0.01 - 4*x*(1 - x))**2
x = np.linspace(-0.2,1.2)
plt.plot(x,func(x))
plt.xlabel("X2")
plt.ylabel("f(x)")
plt.grid()
```
