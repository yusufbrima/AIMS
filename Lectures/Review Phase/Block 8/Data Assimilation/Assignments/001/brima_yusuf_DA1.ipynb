```python
import numpy as np
# from sympy import symbols, solve
from sympy import *
import pylab as p
```

```python
x = symbols('x')
expr = 32*x**3 - 48*x**2 + 17.08*x - 0.54 
sol = solve(expr)
sol
```

```python
f =  lambda x: 32*x**3 - 48*x**2 + 17.08*x - 0.54 
x = np.linspace(-0.2,1.2)
fig = p.figure(1,figsize=(8,6))
p.plot(x,f(x))
# p.plot(sol,'r*',markersize=10)
p.xlabel(r"$X_2$")
p.ylabel(r"$f(x)$")
p.legend([r'$f(x)$'],loc='best')
p.grid()
p.show()
```

```python
f_x =  lambda x: (0.5 - x)**2 + (0.01 - 4*x*(1 - x))**2
x = np.linspace(-0.2,1.2)
fig = p.figure(1,figsize=(8,6))
p.plot(x,f_x(x))
p.legend([r'$J(X_2,Y_3,Y_3)$'],loc='best')
p.xlabel("X2")
p.ylabel("f(x)")
p.grid()
p.show()
```
