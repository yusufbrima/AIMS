```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x0 = np.random.normal(0,1,1)
```

```python
t = np.arange(0,31,1)
```

```python
epsilon = np.random.normal(0,0.1**2)
a = 0.8
```

```python
x = np.zeros(31)
x[0] = x0
for i in range(1,31):
    delta = np.random.normal(0,0.4)
    x[i] = a*x[i-1] + delta
```

```python
plt.plot(t,x)
plt.xlabel('TIME')
plt.ylabel(' $X(t)$ ')
plt.title('\"TRUE\" STATE')
plt.show()
```

```python
def h(t):
    if 1 <= t <= 10 or 21<= t <= 30:
        return 1
    elif 11 <= t <= 20:
        return 0.1
```

```python
y = np.zeros(30)
for i in range(0,30):
    epsilon = np.random.normal(0,0.1)
    y[i] = h(i+1) * x[i+1] +  epsilon
```

```python
plt.plot(t[1:],x[1:])
plt.plot(y, 'r*')
plt.xlabel('TIME')
plt.ylabel(' $X(t)$ ')
plt.legend(['\"TRUE\" STATE', 'DATA'])
plt.show()
```
