## 2.c

```python
x0 = np.random.normal(0,1,1)
tt = np.arange(0,101,1)
alpha  = 0.8
x = np.zeros(101)
q  =  0.5
x[0] = x0
for t in range(1,len(x)):
    delta = np.random.normal(0,q)
    x[t] = alpha *x[t-1] + delta
```

```python
plt.figure(figsize=(10,8))
plt.plot(tt,x,label="True")
plt.xlabel('Time',fontsize=16)
plt.ylabel(' $X(t)$ ',fontsize=16)
plt.title('TRUE STATE',fontsize=16)
plt.xticks(np.arange(0, 101, step=10)) 
plt.legend()
plt.show()
```

## 2.d

```python
xnew = np.concatenate((x[tt[1:40]], x[tt[44:80]],x[tt[84:]]), axis=None)
time = np.concatenate((tt[1:40], tt[44:80],tt[84:]), axis=None)
y = np.zeros(len(time))
tau = 0.1

epsilon = np.random.normal(0,tau,len(time))
y =  xnew  + epsilon
plt.figure(figsize=(10,8))
plt.plot(tt,x)
plt.plot(time,y, 'r^')
plt.xlabel('Time',fontsize=16)
plt.ylabel(rf'$X(t)$ ',fontsize=16)
plt.legend(['TRUE STATE', 'DATA'])
plt.xticks(np.arange(0, 101, step=10)) 
plt.show()
```

## 2.e

```python
def kalman(mu,sigma,r,q,alpha,y,N,nd):
        gain,mu_new,sigma_new=np.zeros(N),np.zeros(N),np.zeros(N)
        mu_old,sigma_old = alpha*mu, q+alpha*sigma*alpha
        for t in range(0,N):
            if (t in nd) == False:
                gain[t] = sigma_old* np.power(sigma_old+r , -1)
                mu_new[t] = mu_old + gain[t]*(y[t]- mu_old) 
                sigma_new[t] = (1 - gain[t])*sigma_old
            else:
                mu_new[t] = mu_old 
                sigma_new[t] = sigma_old 
            mu_old = alpha * mu_new[t]
            sigma_old = q + np.power(alpha,2) *sigma_new[t]
        return sigma_new, gain, mu_new
```

```python
alpha  =  0.75
r = 0.1 #Variance of epsilon_t
q  =  0.5 #Variance of delta_t
mu  =  0
sigma = 1

nd = np.array([40,41,42,43, 80,81,82,83])
y = np.zeros(len(x)-1)
N = len(y)
for i in range (len(y)):
    if (i in nd) == False:
        y[i]=x[i+1]+ np.random.normal(0,tau)
    else:
        y[i] = 0

sigma_new, gain , mu_new =kalman(mu, sigma,np.power(r,2),np.power(q,2),alpha,y,N,nd)
```

```python
plt.figure(figsize=(8,6))
plt.plot(tt,x)
plt.plot(tt[1:],y,"r^")
plt.plot(tt[1:],mu_new,'--g',markersize=4)
plt.legend( [r'True State $x_i$',r'Data $y_i$','Kalman Filter'])
plt.xticks(np.arange(0, 101, step=10)) 
plt.xlabel('Time',fontsize=16)
plt.ylabel("Processes",fontsize=16)
plt.show()
```

### Observations
A close study of the process model $X(t)$ with a standard deviation $\delta = 0.5$ and the data model $Y(t)$ also with a standard deviation of $\tau = 0.1$ firstly gives an indication that the underlying process model has high spatio-temporal variability. However, the data model was able to capture the underlying factors of variation in the process which is indicated by the very low standard deviation of $0.1$ and the close fit of both models. In the areas where there is low data visibility in the data model, we notice a sharp variation in the Kalman Filter estimate as show in the graph. One of the physical interpretation of the low data visibility is intrumentation error which may lead to missing values.

```python
plt.figure(figsize=(10,8))
plt.plot(sigma_new,'b-',markersize=4)
plt.legend( ['Filter error'])
plt.ylabel(r"$\sigma_{t|t}$",fontsize=16)
plt.xlabel('Time',fontsize=16)
plt.title("Filter Estimate Error",fontsize=16)
plt.show()
```

### Filter Estimate Error
The aforementioned graph depicts the filter estimation error, the spikes at the intervals at times 40 through 43 and 80 through 83 generally indicate the missing values in the data model $Y(t)$ which may be due to measurement and /or physical causes.
