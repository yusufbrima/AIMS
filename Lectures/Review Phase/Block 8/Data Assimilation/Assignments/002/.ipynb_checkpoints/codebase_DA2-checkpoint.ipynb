```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
t = np.arange(0,101,1)
x = np.zeros(101)

x[0]=np.random.normal(0,1)
for i in range (1,100):
    x[i] = 0.7*x[i-1] + np.random.normal(0,0.5)
```

```python
plt.plot(t,x)
plt.xlabel('Time', fontsize=20)
plt.ylabel('$x(t)$', fontsize=20)
plt.title("Truth simulated state")
```

```python
t1 = np.concatenate(( np.arange(1,40), np.arange(44,80), np.arange(84,101) ))
r= np.array([40,41,42,43, 80,81,82,83])
    
x1 = np.concatenate(( np.array(x[1:40]),np.array(x[44:80]), np.array(x[84:101])))
y = []
for i in range (len(x1)):
    y.append(x1[i]+ np.random.normal(0,0.1))

plt.plot(t1,x1)
plt.plot(t1,y,"*")
```

```python
def kalman(sig,mu,r,q,alpha,y,N):
        gain=np.zeros(N)
        mu_new=np.zeros(N)
        sig_new=np.zeros(N)
        mu_old=alpha*mu
        sig_old=q+alpha*sig*alpha
        r1= np.array([40,41,42,43, 80,81,82,83])
        for i in range(0,N):
            if (i in r1) == False:
                gain[i]=sig_old*(sig_old+r)**(-1)
                mu_new[i]=mu_old+gain[i]*(y[i]-mu_old) 
                sig_new[i]=(1-gain[i])*sig_old 
            else:
                mu_new[i]=mu_old 
                sig_new[i]=sig_old 
            mu_old=alpha*mu_new[i]
            sig_old=q+alpha**2*sig_new[i]
        return mu_new, sig_new, gain
```

```python
alpha = 0.7
q =0.5
r = 0.1
N = len(x)
y = np.zeros(len(x))
r1= np.array([40,41,42,43, 80,81,82,83])
for i in range (len(x)):
    if (i in r1) == False:
        y[i]=x[i]+ np.random.normal(0,0.1)
    else:
        y[i] = 0
mu_new, sig_new, gain =kalman(1,0,r**2,q**2,alpha,y,N)
```

```python
plt.figure(1)
plt.plot(t,x)
plt.plot(t,y,"*")
plt.plot(t,mu_new,'--g',markersize=4)
plt.legend( ['true state x_i','data y_i','Kalman filter'])
plt.show()
```
