```python
import numpy as np
import matplotlib.pyplot as plt
```

## 2.c

```python
x0 = np.random.normal(0,1,1)
t = np.arange(0,101,1)
alpha  = 0.8
x = np.zeros(101)
q  =  0.5
x[0] = x0
for k in range(1,len(x)):
    delta = np.random.normal(0,q)
    x[k] = alpha *x[k-1] + delta
```

```python
plt.figure(figsize=(10,8))
plt.plot(t,x,label="True")
plt.xlabel('Time')
plt.ylabel(' $X(t)$ ')
plt.title('TRUE STATE')
plt.legend()
plt.show()
```

```python
M_t  =  0.75
T = 100
sigma_0_0  =  1
mu_0_0 =  0 
R_t = 0.1 #Variance of epsilon_t
Q_t  =  0.5 #Variance of delta_t
H_t =  1
sigma = np.zeros(T+1)
mu  = np.zeros(T+1)
k = np.zeros(T+1)
y =  np.random.normal(0.2,0.05,101)
t = 0
for t in range(1, len(sigma)):
    sigma[t- 1]  =  Q_t +  M_t *  sigma_0_0
    mu[t-1]  =  M_t * mu_0_0
    k[t-1] = sigma[t- 1]  * H_t * (   H_t * sigma[t- 1]   * H_t + R_t)**-1
    mu[t]  = mu[t-1] + k[t-1] *( y[t]   -  H_t * mu[t-1] )
    sigma[t]  =  (1 -  k[t-1] * H_t) * sigma[t- 1]
```
